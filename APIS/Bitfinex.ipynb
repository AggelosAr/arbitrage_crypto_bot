```python

# Ratelimit: 90 req/min

def get_tickers_Bitfinex():
    
    try:
        URL = "https://api-pub.bitfinex.com/v2/tickers?symbols=ALL"
        location = "!"
        PARAMS = {'address':location}

        r = requests.get(url = URL, params = PARAMS)

        tickers = r.json()
    except:
        return "ERROR in get_Tickers_Bitfinex"
    temp = []  
    
    if len(tickers[0])== 11:
        
        for i in range(len(tickers)):
            # Pairs here must start with "t" declaring a trade token, we check for that.
            if tickers[i][0][0] == "t":

                temp.append([tickers[i][0][1:]]+tickers[i][1:])
                
        temp= np.array(temp)
        
        ok = temp[:, [0, 7]]
        
        ok = ok.tolist()
        return ok
    else:
        return "ERROR IN get_Tickers_Bitfinex"
    

"""
ticker composition:
BID,                          1
  BID_SIZE,                   2   
  ASK,                        3
  ASK_SIZE,                   4                   
  DAILY_CHANGE,               5
  DAILY_CHANGE_RELATIVE,      6
  LAST_PRICE,                 7
  VOLUME,                     8
  HIGH,                       9
  LOW                         10
"""


def get_Ticker_Bitfinex(coin_pair):
    
    URL = "https://api-pub.bitfinex.com/v2/ticker/"+coin_pair
    location = "!"
    PARAMS = {'address':location}
    
    r = requests.get(url = URL, params = PARAMS)
    
    ticker = r.json()
    
    if len(ticker) == 10:
        return ticker
    else:
        return "ERROR IN get_Ticker_Bitfinex"
    

def get_book_Bitfinex(coin_pair):
    URL = "https://api-pub.bitfinex.com/v2/book/t"+coin_pair+"/R0"
    location = "!"
    PARAMS = {'address':location}
    
    r = requests.get(url = URL, params = PARAMS)
    
    book = r.json()
    
    return book
```
