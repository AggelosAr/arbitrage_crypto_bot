```python
#all_coins = get_currencies_HitBTC()
#all_pairs_kraken = get_pairs_Kraken()

def connect_to_api(function_name):
    
    return eval("get_tickers_" + function_name + "()")


def get_data(markets):
    
    with ThreadPoolExecutor(max_workers=14) as executor:
        # m_market is the name of the market in string format in a seperate 'header' file
        # we need to cast it as a variable and add get_tickers in front of it to 
        # call the appropriate function
        future_to_market = {executor.submit( connect_to_api, m_market ): m_market for m_market in markets}
        
        for future in concurrent.futures.as_completed(future_to_market):
            market = future_to_market[future]
            try:
                data = future.result()
                markets[market] = data
            except Exception as exc:
                print('%r generated an exception: %s' % (url, exc))

                
def cleanup_markets(markets):
    
    to_delete = []
    #Iterate all the markets we found and check for errors.
    for market, value in markets.items():
        if (type(value) != list):
            print(value)
            to_delete.append(str(market))
        else:
            print("Found ", len(value)," pairs in ",  market, ".")

    #In case of error, delete the market from the dictionary.
    for market in to_delete:
        del markets[market]
        print("Deleting : ", market)
        
```

```python
prices_Bitfinex  = { m_Bitfinex[i][0] : {"Bitfinex" : float(m_Bitfinex[i][1])} for i in range(len(m_Bitfinex))}
prices_Binance  = { m_Binance[i][0] : {"Binance" : float(m_Binance[i][1])} for i in range(len(m_Binance))}
prices_FTX2  = { m_FTX2[i][0] : {"FTX2" : float(m_FTX2[i][1])} for i in range(len(m_FTX2))}
#prices_HitBTC  = { m_HitBTC[i][0] : {"HitBTC" : float(m_HitBTC[i][1])} for i in range(len(m_HitBTC))}

prices_BitMart2  = { m_BitMart2[i][0] : {"BitMart2" : float(m_BitMart2[i][1])} for i in range(len(m_BitMart2))}
prices_KuCoin2  = { m_KuCoin2[i][0] : {"KuCoin2" : float(m_KuCoin2[i][1])} for i in range(len(m_KuCoin2))}
prices_BigONE2  = { m_BigONE2[i][0] : {"BigONE2" : float(m_BigONE2[i][1])} for i in range(len(m_BigONE2))}
prices_BtcTurk_PRO  = { m_BtcTurk_PRO[i][0] : {"BtcTurk_PRO" : float(m_BtcTurk_PRO[i][1])} for i in range(len(m_BtcTurk_PRO))}
prices_Bitbank2  = { m_Bitbank2[i][0] : {"Bitbank2" : float(m_Bitbank2[i][1])} for i in range(len(m_Bitbank2))}
prices_B2BX2  = { m_B2BX2[i][0] : {"B2BX2" : float(m_B2BX2[i][1])} for i in range(len(m_B2BX2))}




all_Platforms = [prices_Bitfinex, prices_Binance, prices_FTX2, 
                  prices_BitMart2, prices_KuCoin2, 
                 prices_BigONE2, prices_BtcTurk_PRO, prices_Bitbank2, prices_B2BX2]

# Merge all together 
result = defaultdict(list)
for d in all_Platforms:
    for k, v in d.items():
        result[k].append(v)
        
all_Platforms = result

ok = 0
counter = 0

for k in all_Platforms:
    
    # COIN  --->>>> list of dicitionaries e.x.  'BTCUSD' : [{'Bitfinex': 33954.0}, {'FTX2': 33933.0}]
    
    if len(all_Platforms[k]) > 1:  # List of dictionaries must be > 1 in order for a trade to be possible.
        
        # Merge a list of dicts into a single dict.
        merge = {}
        
        for el in all_Platforms[k]:
            merge.update(el)
            
        min_value = min(merge.values())
        min_name = min(merge, key=merge.get)
        
        max_value = max(merge.values())
        max_name = max(merge, key=merge.get)
        
        dif = np.abs(min_value-max_value)
        
        

            
        if dif >= 1 :
            if (min_name == "HitBTC") or (max_name == "HitBTC"):
                ok += 1
            counter += 1
            #print(k, all_Platforms[k] )
            print(k,  all_Platforms[k],  dif)
            
                  
```
